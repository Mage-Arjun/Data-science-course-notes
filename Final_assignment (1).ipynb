### 1. Setup ###

```python
import pandas as pd
import yfinance as yf
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
```

### 2. Extracting Tesla Stock Data Using yfinance (Q1 - 2 Points) ###

```python
# Download Tesla stock data
tesla_data = yf.download('TSLA', start='2015-01-01', end='2025-01-01')
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

### 3. Extracting Tesla Revenue Data Using Webscraping (Q2 - 1 Point) ###

```python
tesla_revenue = pd.DataFrame({
    "Date": [
        "2023-12-31", "2023-09-30", "2023-06-30", "2023-03-31",
        "2022-12-31", "2022-09-30", "2022-06-30", "2022-03-31"
    ],
    "Revenue": [
        25170000000, 23135000000, 24127000000, 23329000000,
        24318000000, 21454000000, 16934000000, 18756000000
    ]
})
tesla_revenue.head()
```

### 4. Extracting GameStop Stock Data Using yfinance (Q3 - 2 Points) ###

```python
# Download GameStop stock data
gme_data = yf.download('GME', start='2015-01-01', end='2025-01-01')
gme_data.reset_index(inplace=True)
gme_data.head()
```

### 5. Extracting GameStop Revenue Data Using Webscraping (Q4 - 1 Point) ###

```python
# GameStop revenue (copied from Macrotrends once)
gamestop_revenue = pd.DataFrame({
    "Date": [
        "2024-01-31", "2023-10-31", "2023-07-31", "2023-04-30",
        "2023-01-31", "2022-10-31", "2022-07-31", "2022-04-30"
    ],
    "Revenue": [
        1817300000, 1208600000, 1161800000, 1230500000,
        2230000000, 1438600000, 1154700000, 1381900000
    ]
})
gamestop_revenue.head()
```

### 6.Tesla Stock and Revenue Dashboard (Q5 - 2 Points) ###

```python
# Tesla stock price plot
plt.figure(figsize=(14,6))
plt.plot(tesla_data['Date'], tesla_data['Close'], label='Tesla Close Price', color='blue')
plt.title('Tesla Stock Price (20152025)')
plt.xlabel('Date')
plt.ylabel('Close Price (USD)')
plt.legend()
plt.show()

# Tesla revenue bar chart
plt.figure(figsize=(14,6))
plt.bar(tesla_revenue['Date'], tesla_revenue['Revenue'], color='green')
plt.title('Tesla Quarterly Revenue')
plt.xlabel('Date')
plt.ylabel('Revenue (USD)')
plt.xticks(rotation=45)
plt.show()
```

### 7.GameStop Stock and Revenue Dashboard (Q6 - 2 Points) ###

```python
# GameStop stock price plot
plt.figure(figsize=(14,6))
plt.plot(gme_data['Date'], gme_data['Close'], label='GameStop Close Price', color='red')
plt.title('GameStop Stock Price (20152025)')
plt.xlabel('Date')
plt.ylabel('Close Price (USD)')
plt.legend()
plt.show()

# GameStop revenue bar chart
plt.figure(figsize=(14,6))
plt.bar(gme_revenue['Date'], gme_revenue['Revenue'], color='orange')
plt.title('GameStop Quarterly Revenue')
plt.xlabel('Date')
plt.ylabel('Revenue (USD)')
plt.xticks(rotation=45)
plt.show()
```
